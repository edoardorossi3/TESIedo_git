We obtain T50 with bisection

```python
mass=lambda x: (1-math.exp(-x**2/(2*tau**2)))/(1-math.exp(-t_obs**2/(2*tau**2)))
```

We define a function bisection

```python
def bisection(f, t_min, t_max, toll):
    if f(t_min)*f(t_max)>=0:
        return None
    while (t_max-t_min)>toll:
        t_m=(t_max+t_min)/2.0
        if f(t_min)*f(t_m)>0:
            t_min=t_m
            t_max=t_max
        elif f(t_max)*f(t_m)>0:
            t_max=t_m
            t_min=t_min
        elif f(t_m)==0:
            return t_m
    return (t_max+t_min)/2.0
```

```python
mass_50=lambda x: mass(x)-0.5
t50_bis=bisection(mass_50, 0.0, t_obs, 0.001)
print(t50_bis, 't50 derived by bisection')
t50=tau*(-2*math.log(0.5+0.5*math.exp(-0.5*(t_obs**2)/(tau**2))))**0.5
print(mass(t50_bis))
print(mass(t50))
print(t50, 't50 derived analitically')
```

```python
tmw_n=[0.0]*Nbins_t
T75_n=[0.0]*Nbins_t
T25_n=[0.0]*Nbins_t
for i in range (0, Nbins_t):
    T25_n[i]=Tau_n[i]*np.sqrt(-2.0*np.log(0.75+0.25*np.exp(-0.5/Tau_n[i]**2)))
    T75_n[i]=Tau_n[i]*np.sqrt(-2.0*np.log(0.25+0.75*np.exp(-0.5/Tau_n[i]**2)))
    tmw_n[i]=(-np.exp(-0.5/Tau_n[i]**2)+np.sqrt(np.pi/2.0)*Tau_n[i]*math.erf(1.0/(np.sqrt(2.0)*Tau_n[i])))/(1-np.exp(-0.5/Tau_n[i]**2))
plt.plot(Tau_n, tmw_n, label='tmw/t_obs')
plt.plot(Tau_n, T50_n, label='T50/t_obs')
plt.plot(Tau_n, T25_n, label='T25/t_obs')
plt.plot(Tau_n, T75_n, label='T75/t_obs')
plt.minorticks_on()
plt.legend(loc='lower right')
plt.xlabel('tau/t_obs')
fname='time_mass-weighted_plot.png'
plotfile=workdir+fname
plt.savefig(plotfile)
```
