```python
import astropy.io.fits as fits
import numpy as np
import matplotlib.pyplot as plt
```

```python
work_dir='/export/home/extragal/zibetti/no_ownCloud/SteMaGE/data/SEDlibraries/Sandage_v4.1_Zfix_noburst_cb16MILES_1M/'
time_res_file_full='Time_resol_full_Zfix1M_SNR20.fits'
time_res_file='Time_resol_Zfix1M_SNR20_tot.fits'
```

```python
hdul_full=fits.open(work_dir+time_res_file_full)
time_resol_full=hdul_full[1].data
hdul=fits.open(work_dir+time_res_file)
time_res_z62=hdul[1].data['Log_d1090n50_min_z62']
```

```python
time_resol_full=time_resol_full[1:]
np.size(time_resol_full)
```

```python
d_age50=4.15/83
age50_min=6
age50_max=10.15
N_bins=83
```

```python
vector=np.arange(start=age50_min, stop=age50_max, step=d_age50)
vector=np.append(vector, age50_max)
age_50mid=np.zeros(N_bins)

```

```python
for i in range(0, N_bins):
    age_50mid[i]=(vector[i]+vector[i+1])/2
    
```

```python
plt.plot(age_50mid, time_resol_full, label='full')
plt.plot(age_50mid, time_res_z62, label='5 idx+4 color')

plt.xlabel(r'$log_{10}(age_{50})$')
plt.ylabel(r'$log_{10}(\Delta age_n)$')
plt.title(r'$Z=Z_{\odot}, SNR=20$')
plt.legend(loc='upper right')
```
